## Measures of Central Tendencies
Mean, Median and Mode

```
import scipy as sp
org_a = [ 50,70,35,90,120,100,50,60,50] # INR Thousands per month
org_b = [ 60,60,25,100,110,100,120,50,70]
print('Mean sal of org a:',sp.mean(org_a))
print('Mean sal of org b:',sp.mean(org_b))
print('Median sal of org a:',sp.median(org_a))
print('Median sal of org b:',sp.median(org_b))
```

```
import scipy as sp
org_a = [ 50,70,35,90,1000,100,50,60,50] # INR Thousands per month
org_b = [ 60,60,25,100,110,100,120,50,70]
print('Mean sal of org a:',sp.mean(org_a))
print('Mean sal of org b:',sp.mean(org_b))
print('Median sal of org a:',sp.median(org_a))
print('Median sal of org b:',sp.median(org_b))
# Mean is not a recommended measure if data has outliers, extreme values
# Median is better measure in this case
```

```
from scipy import stats
print('Mode sal of org a:',stats.mode(org_a))
print('Mode sal of org b:',stats.mode(org_b))
```

## Measures of Variability

```
train_a = [10,15,12,17,19] # Delay in mins
train_b = [2,30,15,5,21] # Delay in mins
print('Mean delay of train a:',sp.mean(train_a))
print('Mean delay of train b:',sp.mean(train_b))
print('Median delay of train a:',sp.median(train_a))
print('Median delay of train b:',sp.median(train_b))
```

```
# Range = max - min
print("Range of delay of train a:",max(train_a)-min(train_a))
print("Range of delay of train b:",max(train_b)-min(train_b))
```

```
train_a = [10,15,12,17,19] # Delay in mins
train_b = [2,30,15,5,21] # Delay in mins
print("std of delay of train a:",sp.std(train_a))
print("std of delay of train b:",sp.std(train_b))
```
