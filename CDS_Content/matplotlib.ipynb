# MATPLOTLIB

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 5,4
sb.set_style('whitegrid')
```

```python
x = [1,2,3,4,5]
y = [3,5,8,10,14]
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Simple Graph")
plt.plot(x,y)
plt.show()
```

```python
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Simple  Scatter Plot")
plt.scatter(x,y)
plt.show()
```

```python
# bar,barh,pie
plt.pie(y,[0,0,0,0.2,0],labels=['One','Two','Three','Four','Five'])
plt.show()
```

# Object oriented Plotting-4 Steps
* Create a balnk figure object
* Add axes
* Generate plots from your object
* Specify the layout and plotting parameters

```python
x
```

```python
y
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) #Left,bottom ,width ,height(0 to 1)
ax.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([1,0.5,0.7,0.4])#Left,bottom ,width ,height(0 to 1)
sec = ax.secondary_xaxis('top')
ax.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) #Left,bottom ,width ,height(0 to 1)
ax.set_xlim([2,4])
#ax.set_ylim([5,10])
#ax.set_xticks([2,4])
#ax.set_yticks([10,14])
ax.plot(x,y)
plt.show()
```

```python
plt.bar(x,y,width=[0.1,0.5,0.4,0.2,0.6],color=['red','blue','green','cyan','orange'])
```

```python
plt.plot(x,y,linewidth=2,ls='-',marker='p',mew=15,mfc='yellow',color='cyan')
```

```python
fig, ax = plt.subplots(2,3)# sub plot of 2R and 3C
ax[0,0].bar(x,y)
ax[0,1].plot(x,y)
ax[0,2].scatter(x,y)
ax[1,0].pie(x)
ax[1,1].barh(x,y)
ax[1,2].pie(y)
plt.show()
```
