# Transforming Data and Feature Creation

```python
cars = pd.read_csv('mtcars.csv')
cars.head(2)
```

```python
cars.shape
```

```python
#Adding a new record
cars_new = pd.DataFrame(['Tata Nano',25,5,150,120,3.5,2.56,15.43,0,0,4,4])
cars_new
```

```python
d=cars_new.transpose()
d
```

```python
cars.columns
```

```python
d.columns=['car_model', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs','am', 'gear', 'carb']
d
```

```python
cars_up=pd.concat([cars,d],ignore_index=True)
cars_up
```

```python
cars_up1=pd.concat([d,cars])#,ignore_index=True)
cars_up1
```

```python
cars_up1=pd.concat([d,cars],ignore_index=True)
cars_up1
```

```python
cars.loc[cars.mpg>20,'Banglore'] = 1
cars.loc[cars.mpg<=20,'Banglore'] = 0
```

```python
cars
```

```python
cars.loc[(cars.mpg>20) | (cars.am==1),'Good'] = 1
cars.loc[(cars.mpg<=20) | (cars.am==0),'Good'] = 0
cars
```
